# Task 2
## write a program to simulate 1000 tosses of a fair coin (use np.random.binomial)
## Calculate the mean and standard deviation of that sample

```python
n_tosses = 1000
p_heads = 0.5

tosses = np.random.binomial(1, p_heads, size=n_tosses)

mean = np.mean(tosses)
std_dev = np.std(tosses)

print("Sample mean: {:.4f}".format(mean))
print("Sample standard deviation: {:.4f}".format(std_dev))
```

# Task 3
## Produce a scatterplot of y = 0.5*x+e where x has gaussian (0, 5) and e has gaussian (0, 1) distributions 
### use numpy.random.normal to generate gaussian distribution

```python
n = 1000

#generate x and e as Gaussian distributions
x = np.random.normal(loc=0, scale=5, size=n)
e = np.random.normal(loc=0, scale=1, size=n)

#calculate y using the formula y = 0.5*x + e
y = 0.5*x + e

#create a scatterplot of x vs. y
plt.scatter(x, y, alpha=0.5)
plt.title("Scatterplot of y = 0.5*x + e")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
