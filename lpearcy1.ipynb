# Task 1
## write a program to produce Fibonacci numbers up to 1000000

```python
fib_list = [1,1]
while fib_list[-2]+fib_list[-1] < 1000000:
    fib_list.append(fib_list[-2]+fib_list[-1])
    
print(fib_list)
```

# Task 2
## write a program to simulate 1000 tosses of a fair coin (use np.random.binomial)
## Calculate the mean and standard deviation of that sample

```python
sample = np.random.binomial(n = 1, p = 0.5, size = 1000)
mu = np.mean(sample)
sigma = np.std(sample)

print(mu, sigma)
```

# Task 3
## Produce a scatterplot of y = 0.5*x+e where x has gaussian (0, 5) and e has gaussian (0, 1) distributions 
### use numpy.random.normal to generate gaussian distribution

```python
x = np.random.normal(0,5, size = 1000)
e = np.random.normal(0,1, size = 1000)

plt.scatter(x, 0.5*x+e)
plt.xlabel("x", fontsize = 14, marker = '.')
plt.ylabel("y", fontsize = 14)
plt.show()

plt.scatter(e, 0.5*x+e)
plt.xlabel("e", fontsize = 14)
plt.ylabel("y", fontsize = 14)
plt.show()
```
